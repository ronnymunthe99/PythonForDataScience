# Explore your data visually with Seaborn

Seaborn adalah sebuah pustaka yang memperbaiki fungsionalitas Matplotlib, menggantikan beberapa pengaturan dan fungsi default, dan menambahkan fitur baru.

Seaborn diciptakan untuk memperbaiki tiga kekurangan dari Matplotlib. Seperti yang berdiri sendiri, Matplotlib:



1.   Tidak dapat menghasilkan grafik berkualitas estetika tinggi  
2.   Tidak memiliki fungsionalitas untuk membuat analisis statistik yang rumit dengan mudah.
3. Fitur fungsi yang tidak dirancang untuk berinteraksi dengan Pandas Dataframes 


```python
# Plotting sederhana menggunakan Seaborn
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
sns.set()
x = np.linspace(0, 10, 500)
y = np.random.randn(500)
plt.plot(x,y)
```

```python
sns.distplot(y, kde=True);
```

Seaborn juga menyediakan fungsi untuk menghasilkan plot yang berguna untuk analisis statistik. Misalnya, distplot memungkinkan kita untuk tidak hanya melihat histogram dari sampel, tetapi juga memperkirakan distribusi dari mana sampel berasal.

```python
# Load dataset iris
iris = sns.load_dataset("iris")
iris.head()
```

```python
sns.pairplot(iris, hue='species', height=2.5);
```

Sebelumnya, kita harus tahu data iris adalah data yang mengklasifikasi bunga iris ke dalam 3 jenis bunga yaitu, setosa, versicolor, dan virginica. Masing-masing diklasifikasi dengan 4 kriteria/fitur/ciri yaitu sepal-length, sepal-width, petal-length, dan petal-width. 


Secara esensial, setiap variabel (sepal-length, sepal-width, petal-length, dan petal-width. ) diwakili pada sumbu X dan Y.

Histogram terjadi ketika variabel bersimpangan dengan diri mereka sendiri, misalnya, ketika sepal-length pada sumbu X bersimpangan dengan petal-length pada sumbu Y. Tentu saja, data telah dipisahkan berdasarkan spesies, di mana biru mewakili setosa, hijau mewakili versicolor, dan merah mewakili virginica.

Elemen lain sebenarnya adalah plot bivariate dari variabel. Misalnya, plot pertama di kolom ketiga mewakili plot bivariate dari sepal-length dan petal-length. Tipe output ini sangat berguna untuk menarik kesimpulan. Misalnya, dalam plot yang sama, dapat dengan mudah mengamati hubungan linier antara sepal-length dan petal-length!

```python
with sns.axes_style('white'):
    sns.jointplot("petal_length", "petal_width", data=iris, kind='reg')
```

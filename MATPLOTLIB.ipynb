# Create Graphs With Matplotlib

```python
# Setup untuk working environment
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

# Generate Simple Graphs

```python
# Mem-plotting sebuah fungsi dengan figure kosong
fig = plt.figure()
ax = plt.axes()
```

Variabel "fig" merupakan varibel yang berisi semua objek (axes, lable, data, dll.). Axes ini merupakan grid yang dapat dilihat di atas, yang akan berisi data grafik

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x));
```

Disini dilakukan plotting sederhana dengan menggunakan **Sin**.
Walaupun sudah terbentuk sebuah grafik, tetapi grafik ini masih belum cukup baik untuk digunakan, karena ada banyak kriteria yang perlu diubah, diantaranya:


*   Ukuran Font
*   Warna Grafik

*   Style Garis 
*   Bagian Putih yang Berlebihan

Dan ada juga beberapa elemen fungsi yang dapat digunakan diantaranya:


*   Judul

*   Legend
*   Label untuk Sumbu








```python
# Mengubah ukuran font default
plt.rcParams.update({'font.size': 15})

fig = plt.figure()
ax = plt.axes()

# linestyle 'solid' yang warnanya ditentukan oleh namanya
plt.plot(x, np.sin(x - 0), color='blue', linestyle='solid', label='bleu')
# Nama warna yang disingkat bisa pakai 'dashed'
plt.plot(x, np.sin(x - 1), color='g', linestyle='dashed', label='vert')
# Grayscale antara 0 dan 1, dashes and dots atau 'dashdot'
plt.plot(x, np.sin(x - 2), color='0.75', linestyle='dashdot', label='gris')
# Warna RGB, 'dotted'
plt.plot(x, np.sin(x - 3), color='#FF0000', linestyle='dotted', label='rouge')
# Batas sumbu. Bisa dicoba 'tight' dan 'equal' untuk melihat efeknya
plt.axis([-1, 11, -1.5, 1.5]);
# Label
plt.title("Example of a graph")
# Legend is digenerate dari argument 'label' of 'plot'
# 'loc' menspesifikkan penempatan legenda.
plt.legend(loc='lower left');
# Nama sumbu(Axis)
ax = ax.set(xlabel='x', ylabel='sin(x)')
```

# Visualize Uncertainity

Aspek lain dari plotting yang harus kita ketahui adalah tingkat ketidakpastian yang terkait dengan perkiraan. Sangat penting untuk memperhitungkan hal ini tidak hanya saat menganalisis data, tetapi juga saat merepresentasikannya.

Untuk memvisualisasikan uncertainity, atau dengan kata lain, untuk mewakili seberapa yakin kita dalam perkiraan kita, kita akan menggunakan error bar.

## Data Diskrit

Dalam kasus data diskrit, sering digunakan batas error untuk mewakili ketidakpastian yang melekat pada nilai setiap titik. Seringkali, panjang batang sesuai dengan standar deviasi dari observasi empiris.

```python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)
plt.errorbar(x, y, yerr=dy, fmt='.k');
```

Errorbar membutuhkan koordinat-x, koordinat-y, dan panjang masing-masing batang (satu batang per titik) yerr sebagai argumen. Pada error bar kali ini, ditambahkan argumen fmt 

```python
plt.errorbar(x, y, yerr=dy, fmt='o', color='black', 
    ecolor='lightgray', elinewidth=3, capsize=0);
```

Errorbar memungkinkan untuk memilih warna (di sini hitam) dan bentuk tanda pada grafik dengan cara yang sangat efisien. Errorbar juga memungkinkan untuk lebih menyesuaikan tampilan grafik.

# Customize Plots

```python
print(plt.style.available[:6])
# Mengubah ukuran figure
fig = plt.figure(figsize=(12,8))
for i in range(6):    
    # menambahkan subplot
    fig.add_subplot(3,2,i+1)    
    plt.style.use(plt.style.available[i])    
    plt.plot(x, y)        
    # plotting
    plt.text(s=plt.style.available[i], x=5, y=2, color='red')
```

Argumen pertama dari fungsi add_subplot adalah jumlah baris pada tabel grafik (di sini 3). Kedua adalah jumlah kolom (di sini 2). Ketiga adalah jumlah grafik yang sebenarnya, di antara grafik dalam tabel ini, yang ingin kita gambar (di sini 1).

**Untuk alasan tertentu, subgrafik diberi dari nomor 1 bukan dari 0, sehingga grafik kiri atas adalah grafik nomor 1.**

```python
x = np.random.randn(1000)

plt.style.use('classic')
fig=plt.figure(figsize=(5,3))
ax = plt.axes(facecolor='#E6E6E6')
# Menampilkan tanda di bawah sumbu
ax.set_axisbelow(True)
# frame putih
plt.grid(color='w', linestyle='solid')

# Sembunyikan frame
for spine in ax.spines.values():    
    spine.set_visible(False)
    
# Sembunyikan tanda di atas dan kanan
ax.xaxis.tick_bottom()
ax.yaxis.tick_left()

# Menentukan tanda, dan merotasi
marqueurs = [-3, -2, -1, 0, 1, 2, 3]
xtick_labels = ['A', 'B', 'C', 'D', 'E', 'F']
plt.xticks(marqueurs, xtick_labels, rotation=30)

# Mengubah warna tanda
ax.tick_params(colors='gray', direction='out')
for tick in ax.get_xticklabels():    
    tick.set_color('gray')
for tick in ax.get_yticklabels():    
    tick.set_color('gray')    
    
# Mengubah warna siku
ax.hist(x, edgecolor='#E6E6E6', color='#EE6666');
```

Untuk menampilkan tanda pada bawah sumbu, gunakan perintah ax.set_xticks(x) dan ax.set_yticks(y), di mana x dan y adalah lokasi tanda pada sumbu-x dan sumbu-y. Ini akan menampilkan tanda pada bawah sumbu. Anda juga dapat mengatur label tanda dengan menggunakan perintah ax.set_xticklabels(labels_x) dan ax.set_yticklabels(labels_y), di mana labels_x dan labels_y adalah label tanda yang diinginkan.
